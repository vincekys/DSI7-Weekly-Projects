# DATA CLEANING

```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import re
import pickle

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

sns.set_style('ticks')
plt.style.use('seaborn')

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
jobs = pd.read_csv('./Data/job_data_v2.csv')
```

### Remove leading and trailing whitespaces. Drop duplicates

```python
# remove whitespaces
jobs = jobs.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# set job_id as index
jobs.set_index('job_id', inplace=True)

# drop duplicate jobs
jobs.drop_duplicates(keep='first', inplace=True)

jobs.head()
```

```python
jobs.shape
```

# Data Cleaning:

### Salary
- expand salary_range to min and max salary
- convert all to monthly salaries
- convert to float
- find the median salary per job
- calculate median salary across all jobs and classify them as either low or high salary

```python
# replace 'Salary undisclosed' with blanks
jobs['salary_range'].replace('Salary undisclosed', np.nan, inplace=True)

# replace '$' with blank
jobs['salary_range'] = jobs['salary_range'].str.replace('$', '')

# replace 'to' with space
jobs['salary_range'] = jobs['salary_range'].str.replace('to', ' ')

# replace comma with blank
jobs['salary_range'] = jobs['salary_range'].str.replace(',', '')
```

```python
# split ' '
jobs[['min_salary', 'max_salary']] = jobs['salary_range'].str.split(expand=True)

# convert min_salary and max_salary to numeric
jobs['min_salary'] = jobs['min_salary'].astype(float)
jobs['max_salary'] = jobs['max_salary'].astype(float)

# change all salary to monthly
jobs['min_salary'] = np.where(jobs['salary_freq'] == 'Annually', jobs['min_salary'] / 12, jobs['min_salary'])
jobs['max_salary'] = np.where(jobs['salary_freq'] == 'Annually', jobs['max_salary'] / 12, jobs['max_salary'])
```

```python
# remove those without min_salary or max_salary
jobs = jobs[jobs['min_salary'].notnull()]
jobs = jobs[jobs['max_salary'].notnull()]
```

```python
# calculate median_salary of each job based on their min and max salary
jobs['median_salary'] = jobs[['min_salary', 'max_salary']].median(axis=1)
```

```python
# check the distribution of median_salary
sns.distplot(jobs['median_salary'], kde=False)
plt.title('distribution of median_salary')
plt.show()
```

```python
# drop the extreme outliers of min_salary > 20k
jobs = jobs[jobs['min_salary'] < 20000]
print(jobs.shape)
```

```python
# drop columns that we are not going to use
jobs.drop(columns=['salary_freq', 'min_salary', 'max_salary', 'salary_range'], inplace=True)
```

```python
# what is the median of median_salary?
print('Median Salary: ${}'.format(jobs['median_salary'].median()))
```

```python
# use $6750 as the threshold. jobs above $6750 would be classified as high salary
jobs['high_salary'] = [1 if x > 6750 else 0 for x in jobs['median_salary']]
jobs.drop('median_salary', axis=1, inplace=True)
```

### job_description and requirements:
- remove the leading 'Requirements' and 'Roles & Responsibilities'
- drop job description as important word features should lie in requirements
- remove null requirements

```python
# remove the leading 'Requirements' and 'Roles & Responsibilities'
jobs['requirements'] = jobs['requirements'].str[12:]
jobs['job_description'] = jobs['job_description'].str[24:]

# drop job_description
jobs.drop('job_description', axis=1, inplace=True)

# drop those without requirements
jobs = jobs[jobs['requirements'].notnull()]
```

### requirements

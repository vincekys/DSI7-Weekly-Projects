# Project 4 - Question 2

### What features distinguish a data analyst & scientist role vs other roles (data engineer etc)

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, roc_auc_score, precision_score
from sklearn.feature_selection import RFECV, SelectKBest
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from imblearn.over_sampling import SMOTE

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')
plt.style.use('seaborn')
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
jobs_df2 = pd.read_pickle('./Data/jobs_df2.pkl')
jobs_df2.head(3)
```

## Oversampling required later on to fix our imbalanced target variable

```python
pd.Series(jobs_df2['data_analyst_scientist']).value_counts().plot('bar')
plt.title('Proportion of data_analyst_scientist')
plt.show()
print(jobs_df2['data_analyst_scientist'].value_counts())
```

## Apply TfidfVectorizer on requirements

```python
tfidf = TfidfVectorizer(stop_words='english', ngram_range=(1,2), max_df=0.95, min_df=0.1)

tfidf_requirements = tfidf.fit_transform(jobs_df2['requirements'])
```

```python
requirements_vec = pd.DataFrame(tfidf_requirements.todense(), columns=tfidf.get_feature_names())
```

```python
requirements_vec.head()
```

```python
df = pd.concat([jobs_df2['data_analyst_scientist'], requirements_vec], axis=1)
```

```python
df.head()
```

```python
X = df.drop('data_analyst_scientist', axis=1)
y = df['data_analyst_scientist']
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify=y, random_state=50)

print('X_train:', X_train.shape)
print('y_train:', y_train.shape)
print('X_test:', X_test.shape)
print('X_test:', y_test.shape)
```

## Logistic Regression

```python
logreg = LogisticRegression()
sm = SMOTE(random_state=99, n_jobs=-1)
X_smote, y_smote = sm.fit_sample(X_train, y_train)
logreg.fit(X_smote,y_smote)
y_pred = logreg.predict_proba(X_test)[:,1]
print(roc_auc_score(y_test, y_pred))
```

## Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB()
sm = SMOTE(random_state=99, n_jobs=-1)
X_smote, y_smote = sm.fit_sample(X_train, y_train)
mnb.fit(X_smote,y_smote)
y_pred = mnb.predict_proba(X_test)[:,1]
print(roc_auc_score(y_test, y_pred))
```

## RandomForestClassifier

```python
rfc = RandomForestClassifier()
sm = SMOTE(random_state=99, n_jobs=-1)
X_smote, y_smote = sm.fit_sample(X_train, y_train)
rfc.fit(X_smote,y_smote)
y_pred = rfc.predict_proba(X_test)[:,1]
print(roc_auc_score(y_test, y_pred))
```

## Feature Importance Using Logistic Regression

```python
feature_importance = abs(logreg.coef_[0])
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5

featfig = plt.figure(figsize=(20,30))
featax = featfig.add_subplot(1, 1, 1)
featax.barh(pos, feature_importance[sorted_idx], align='center')
featax.set_yticks(pos)
featax.set_yticklabels(np.array(X_train.columns)[sorted_idx], fontsize=12)
featax.set_xlabel('Absolute Relative Feature Importance')

plt.title('Absolute Feature Importance')
plt.tight_layout()   
plt.show()
```

```python
feature_importance = logreg.coef_[0]
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5

featfig = plt.figure(figsize=(20,30))
featax = featfig.add_subplot(1, 1, 1)
featax.barh(pos, feature_importance[sorted_idx], align='center')
featax.set_yticks(pos)
featax.set_yticklabels(np.array(X_train.columns)[sorted_idx], fontsize=12)
featax.set_xlabel('Relative Feature Importance')

plt.title('Feature Importance')
plt.tight_layout()   
plt.show()
```

## 1. Estimating the value of homes from fixed characteristics.

---

Your superiors have outlined this year's strategy for the company:
1. Develop an algorithm to reliably estimate the value of residential houses based on *fixed* characteristics.
2. Identify characteristics of houses that the company can cost-effectively change/renovate with their construction team.
3. Evaluate the mean dollar value of different renovations.

Then we can use that to buy houses that are likely to sell for more than the cost of the purchase plus renovations.

Your first job is to tackle #1. You have a dataset of housing sale data with a huge amount of features identifying different aspects of the house. The full description of the data features can be found in a separate file:

- Roof and exterior features
- "Quality" metrics, such as kitchen quality

- Train a model on pre-2010 data and evaluate its performance on the 2010 houses.

```python
# create dummy variables for categorical features
df_fixed = pd.get_dummies(df_fixed)
df_fixed.shape
```

```python
# split the YrSold out before bringing it back later
year_sold = pd.DataFrame(df_fixed['YrSold'])
df_fixed = df_fixed.drop(columns='YrSold')
year_sold.head()
```

```python
df_fixed.head()
```

```python
# standardise the whole df
df_fixed = (df_fixed - df_fixed.mean()) / df_fixed.std()
df_fixed.head()
```

```python
# merge back the YrSold
df_fixed = df_fixed.join(year_sold)
df_fixed.head()
```

```python
# SPLIT THE DATA INTO PRE AND POST 2010
df_fixed_2010 = df_fixed[df_fixed['YrSold'] == 2010].drop(columns='YrSold')
df_fixed = df_fixed[df_fixed['YrSold'] != 2010].drop(columns='YrSold')
```

```python
# create training data to contain houses sold before 2010 (YrSold)
# create test data to contain houses sold in 2010 (YrSold)
train = df_fixed
test = df_fixed_2010

train = shuffle(train, random_state=42)

train.head()
```

```python
X_train = train.drop(columns='SalePrice')
y_train = train['SalePrice'].values
```

```python
X_test = test.drop(columns='SalePrice')
y_test = test['SalePrice'].values
```

```python
X_train.shape
```

# MODEL: Linear Reg

```python
lreg = LinearRegression()
lreg = lreg.fit(X_train, y_train)
scores = cross_val_score(lreg, X_train, y_train, cv=10)
print(scores)
scores.mean()

# OVERFITTING LIKE CRAZY
```

# MODEL: Ridge Reg

```python
# find optimal alpha for ridge
ridge_alphas = np.logspace(-2, 7, 50)

optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(X_train, y_train)

print(optimal_ridge.alpha_)
```

```python
ridge = Ridge(alpha=optimal_ridge.alpha_)

ridge_scores = cross_val_score(ridge, X_train, y_train, cv=10)

print(ridge_scores)
print(np.mean(ridge_scores))

# MUCH BETTER R2 WITH RIDGE! 0.774
```

# MODEL: Lasso Reg

```python
# find optimal lasso
optimal_lasso = LassoCV(cv=10, verbose=1)
optimal_lasso.fit(X_train, y_train)

print(optimal_lasso.alpha_)
```

```python
lasso = Lasso(alpha=optimal_lasso.alpha_)

lasso_scores = cross_val_score(lasso, X_train, y_train, cv=10)

print(lasso_scores)
print(np.mean(lasso_scores))

# ALSO VERY GOOD R2 OF 0.768
```

```python
lasso.fit(X_train, y_train)
```

```python
# see how well lasso predicts the testing data
y_pred = lasso.predict(X_test)
y_test
```
